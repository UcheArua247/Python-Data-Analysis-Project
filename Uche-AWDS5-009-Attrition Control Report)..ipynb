#### 2. What Type of Employees Are Leaving?

Based on the exploratory data analysis comparing employees who left the organization with those who remained, the following employee profile is most associated with attrition at Xenos:

Key Characteristics of Employees Who Are Leaving

1. Low Satisfaction Levels
Employees who left consistently show significantly lower satisfaction levels compared to those who stayed. This indicates that dissatisfaction is a strong driver of attrition and serves as an early warning indicator of potential employee exits.

2. High Workload
Employees who left tend to work longer average monthly hours, often exceeding the typical workload of retained employees. This suggests that excessive workload and possible burnout contribute significantly to employee turnover.

3. Project Overload or Underutilization
Attrition is higher among employees handling either too many projects (5–6 projects) or too few projects (2 projects). This implies that both work overload and lack of engagement can increase the likelihood of employees leaving.

4. Lack of Career Progression
A large proportion of employees who left had not received any promotion in the last five years. This highlights limited career advancement opportunities as a major factor influencing employee exits.

5. Lower Salary Levels
Employees earning low salaries are more likely to leave the organization compared to those in medium or high salary brackets. Compensation appears to play a crucial role in retention, especially when combined with high workload or long tenure.

6. Departmental Concentration
Attrition is more pronounced in certain departments, particularly sales, technical, and support roles. These departments may experience higher pressure, performance demands, or workload imbalance.

Summary Insight

Overall, employees who leave Xenos are typically dissatisfied, overworked, under-rewarded, and experience limited career progression. Importantly, many leavers still demonstrate reasonable or high performance evaluations, suggesting that attrition is driven more by disengagement and burnout rather than poor performance.

#### 3. Employees at Risk of Leaving

Based on patterns observed among employees who previously left, a high-risk employee group was identified using the following characteristics:
- Low satisfaction level
- High average monthly working hours
- No promotion in the last five years
- Low salary level
- Long years of service

Employees matching this profile closely resemble past leavers and should be prioritized for retention strategies.

#### 4. Key Drivers of Attrition
##### Attrition Analysis – Who Is Leaving?

To understand employee attrition, employees who left were compared with those who remained across key variables such as satisfaction, workload, and number of projects.

The analysis reveals that employees with low satisfaction levels and those experiencing excessive workloads are more likely to leave the organization. Interestingly, some employees with high performance evaluations still exit the company, suggesting disengagement rather than poor performance.

#### 5. Business Recommendations
Based on the insights derived from this analysis, the following recommendations are proposed:

1. Implement workload balancing strategies for employees handling excessive projects.
2. Introduce clear and transparent promotion pathways to improve career progression.
3. Review compensation structures, particularly for low-salary employees with long tenure.
4. Conduct regular employee satisfaction surveys to detect disengagement early.
5. Apply department-specific retention strategies, as attrition patterns vary across departments.
